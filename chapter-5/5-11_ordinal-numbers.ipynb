# 5-11: Ordinal Numbers

Ordinal numbers indicate their position in a list, such as 1st or 2nd. Most ordinal numbers end in _th_, except 1, 2, and 3.

- Store the numbers 1 through 9 in a list.
- Loop through the list.
 - Use an `if-elif-else` chain inside the loop to print the proper ordinal ending for each number. Your output should read `"1st 2nd 3rd 4th 5th 6th 7th 8th 9th"`, and each result should be on a separate line.


```python
# storing numbers 1-9
nums = list(range(1,10))
print(nums)
```

```python
# practice looping thru numbers
for num in nums:
    print(num)
```

```python
# change number format
for num in nums:
    if num == 1:
        print(f"{num}st")
    elif num == 2:
        print(f"{num}nd")
    elif num == 3:
        print(f'{num}rd')
    else:
        print(f'{num}th')
```
