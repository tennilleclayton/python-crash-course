# 6-3: Glossary

A Python dictionary can be used to model an actual dictionary. However, to avoid confusion, let's call it a glossary.

- Think of five programming words you've learned about in the previous chapters. Use these words as the keys in your glossary, and store their meanings as values.
- Print each word and its meaning as neatly formatted output. You might print the word followed by a colon and then its meaning, or print the word on one line and then print its meaning indented on a second line. Use the newline character (`'\n'`) to insert a blank line between each word-meaning pair in your output.


```python
# Glossary dictionary
glossary = {
    'variables': 'containers for storing data values.', 
    'strings': 'sequence of characters written inside quotes.',
    'tuple': 'an ordered, and unchangeable, collection',
    'lists': 'an ordered, and changeable, collection',
    'comments': 'code lines that will not be executed',
    }
```

```python
# Print word and meaning

word = glossary['variables'].capitalize()
print(f"Variables - {word}")

word = glossary['strings'].capitalize()
print(f"\nStrings - {word}")

word = glossary['tuple'].capitalize()
print(f"\nTuple - {word}")

word = glossary['lists'].capitalize()
print(f"\nLists - {word}")

word = glossary['comments'].capitalize()
print(f"\nComments - {word}")
```
