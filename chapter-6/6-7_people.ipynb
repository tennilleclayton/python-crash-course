# 6-7: People
Start with the program you wrote for Exercise 6-1 (page 98). Make two new dictionaries representing different people, and store all three dictionaries in a list called `people`. Loop through your list of people. As you loop through the list, print everything you know about each person.

```python
# Dictionaries
person_0 = {
    'first_name': 'jon', 
    'last_name': 'johnson', 
    'age': 25, 
    'city': 'atlanta',
    }

person_1 = {
    'first_name': 'mary', 
    'last_name': 'williams', 
    'age': 65, 
    'city': 'augusta',
    }

person_2 = {
    'first_name': 'ariel', 
    'last_name': 'harris', 
    'age': 20, 
    'city': 'rockville',
    }
```

```python
# People list
people = []

people.append(person_0)
people.append(person_1)
people.append(person_2)

# print(people)

for person in people:
    name = f"{person['first_name'].title()} {person['last_name'].title()}"
    age = f"{person['age']}"
    city = f"{person['city'].title()}"
    print(f"{name} is {age} years old from {city}")
```
