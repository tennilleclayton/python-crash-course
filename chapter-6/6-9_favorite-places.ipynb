# 6-9: Favorite Places
Make a dictionary called `favorite_places`. Think of three people names to use as keys in the dictionary, and store one to three favorite places for each person. To make this exericse a bit more interesting, ask some friends to name a few of their favorite places. Loop through the dictionary, and print each person's name and their favorite places.

```python
# Dictionary
favorite_places = {
    'frank': ['disney world', 'home', 'grand canyon'], 
    'mary': ['japan', 'london', 'new york city'], 
    'kevin': ['library', 'los vegas'],
    }
```

```python
# Print each persons favorite place
for person, place in favorite_places.items():
    name = person.title()

    print(f"\n{name}'s favorite places:")
    for location in place:
        print(f"- {location.title()}")
```
