# 6-10: Favorite Numbers
Modify your program from Exercise 6-2 (page 98) so each person can have more than one favorite number. Then print each person's name along with their favorite numbers.

```python
# Program from Exercise 6-2
favorite_numbers = {
    'becky': 10, 
    'george': 7,
    'henry': 12,
    'victora': 13,
    'mark': 2,
    }

print(f"Becky's favorite number is {favorite_numbers['becky']}.")
print(f"George's favorite number is {favorite_numbers['george']}.")
print(f"Henry's favorite number is {favorite_numbers['henry']}.")
print(f"Mark's favorite number is {favorite_numbers['mark']}.")
print(f"Victoria's favorite number is {favorite_numbers['victora']}.")
```

```python
# Append each person favorite number
favorite_numbers['becky'] = [10, 20, 7]
favorite_numbers['george'] = [2, 7]
favorite_numbers['henry'] = [12, 10]
favorite_numbers['victora'] = [3, 6]
favorite_numbers['mark'] = [1, 6]

print(favorite_numbers)
```

```python
# Print each person with their numbers
for name, numbers in favorite_numbers.items():
    print(f"\n{name.title()}'s favorite numbers are:")

    for number in numbers:
        print(f"- {number}")
```
