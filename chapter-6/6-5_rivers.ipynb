# 6-5: Rivers
Make a dictionary containing three major rivers and the country each river runs through. One key-value pair might be `'nile': 'egypt'`.
- Use a loop to print a sentence about each river, such as _The Nile runs through Egypt_.
- Use a loop to print the name of each river included in the dictionary.
- Use a loop to print the name of each country included in the dictionary.


```python
# River dictionary
rivers = {'nile': 'egypt', 'amazon': 'brazil', 'mississippi': 'united states'}

# Loop to print statement
for river, country in rivers.items():
    print(f"The {river.title()} runs through {country.title()}.")
```

```python
# Print Rivers
print("The following rivers are included in this data set:")
for river in rivers.keys():
    print(f"- {river.title()}")
```

```python
# Print Country
print("The following countries are included in this data set:")
for country in rivers:
    print(f"- {rivers[country].title()}")
```
