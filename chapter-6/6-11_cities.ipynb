# 6-11: Cities
Make a dictionary called `cities`. Use the names of three cities as keys in your dictionary. Create a dictionary of information about each city and include the country that the city is in, its approximate population, and one fact about that city. The keys for each city's dictionary should be something like `country`, `population`, and `fact`. Print the name of each city and all of the information you have stored about it.

```python
# Create dictionary
cities = {
    'lithonia': {
        'country': 'united states', 
        'population': 2_600, 
        'fact': 'famously known as the City of Stone'
        }, 
    'london': {
        'country': 'england', 
        'population': 9_000_000, 
        'fact': 'founded by the Romans around 43 CE'
        }, 
    'lagos': {
        'country': 'nigeria', 
        'population': 21_000_000, 
        'fact': 'the largest and fastest-growing city in Africa'
        }, 
    }
```

```python
# Print dictionary information
#for city in cities.keys():
for city, city_info in cities.items():
    #for city, city_info in cities.items():
    loc = city_info['country']
    pop = city_info['population']
    history = city_info['fact']

    print(f"\n{city.upper()} is located in {loc}")
    print(f"  It has a population of about {pop}")
    print(f"  Facts: {history.title()}")
```
