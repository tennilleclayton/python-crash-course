# 7-1: Rental Car
Write a program that asks the user what kind of rental car they would like. Print a message about that car, such as "Let me see if I can find you a Subaru".

```python
question = "What kind of rental car you want? "
prompt = input(question)
print(f"Let me see if I can find you a {prompt}.")
```

```python
car = input("What brand of car you would like to rent? ")
print(f"Let me see if I can find a {car.title()}.")
```
