#### 7-4: Pizza Toppings
Write a loop that prompts the user to enter a series of pizza toppings until they enter a quit value. As they enter each topping, print a message saying you'll add that topping to their pizza.

```python
prompt = f"\nWhat a topping do you want on your pizza: "
prompt += f"\nEnter 'quit' when you are done adding your pizza toppings."

message = ""

# While loop
while message != 'quit':
    print(prompt)
    message = input(prompt)
    print(f"- {message.title()} are added to your pizza.")
```
