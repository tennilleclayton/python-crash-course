# 7-9: No Pastrami
Using the list `sandwich_orders` from Exercise 7-8, make sure the sandwich `'pastrami'` appears in the list at least three times. Add code near the beginning of your program to print a message saying the deli has run out of pastrami, and then use a while loop to remove all occurences of `'pastrami'` from `sandwich_orders`. Make sure no pastrami sandwiches end up in `finished_sandiches`.

```python
# List of from Exercise 7-8 with 'pastrami' added
sandwich_orders = [
    'pastrami', 'peanut butter jelly', 
    'pastrami', 'ham', 'pastrami', 
    'turkey', 'ruben', 'chicken salad'
    ]
finished_sandwiches = []

message = "Sorry, we are out of Pastrami."
```

```python
print("I will like the following sandwiches:")
for sandwich in sandwich_orders:
    print(f"- {sandwich.title()}")
```

```python
print(message)

# Removes Out of Stock item
while 'pastrami' in sandwich_orders:
    sandwich_orders.remove('pastrami')
```

```python
# Making sandwiches
while sandwich_orders:
    current_sandwich = sandwich_orders.pop()

    print(f"I made your {current_sandwich} sandwich.")
    finished_sandwiches.append(current_sandwich)

# Display all sandwiches have need made
print(f"\nThe following sandwiches were made:")
for sandwich in finished_sandwiches:
    print(f"- {sandwich.title()}")
```
